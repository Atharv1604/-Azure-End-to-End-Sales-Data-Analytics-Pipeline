##Doing transformation for all other tables [Changing Column names ]

```
table_names=[i.name.split('/')[0] for i in dbutils.fs.ls("/mnt/adventureworks-lt-2022/Silver/SalesLT")]
```

```
table_names
```

```
from pyspark.sql.functions import *

for i in table_names:
    df = spark.read.format("delta").load("/mnt/adventureworks-lt-2022/Silver/SalesLT/"+i)
    columns_name=df.columns
    for j in columns_name:
        new_col_name="".join(["_"+ char if char.isupper() and j[k-1].islower() else char for k, char in enumerate(j) ]).lstrip("_")
        df=df.withColumnRenamed(j,new_col_name)
    df.write.format("delta").mode("overwrite").save("/mnt/adventureworks-lt-2022/gold/SalesLT"+"/"+i)
```
